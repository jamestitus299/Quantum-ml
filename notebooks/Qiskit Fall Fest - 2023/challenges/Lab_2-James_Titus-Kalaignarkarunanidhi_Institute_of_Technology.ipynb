SAMPLE - 1 SINGLE QUBIT CIRCUIT
</div>

```python
# necessary imports
from qiskit import QuantumRegister, QuantumCircuit, Aer, execute
from qiskit.visualization import *
```

```python
qc = QuantumCircuit(1)    #Create a quantum circuit with one qubit.
```

```python
qc.x(0)                   # Apply X gate to qubit 0
```

```python
qc.draw('mpl')            #Visualize the quantum circuit using Matplotlib.
```

```python
backend = Aer.get_backend('statevector_simulator')    #Select the Aer backend for simulating the quantum statevector.
```

```python
job=execute(qc,backend)   # Execute the quantum circuit 'qc' on the chosen backend
result=job.result()       # Retrieve the result of the executed quantum circuit.
result.get_statevector()  # Obtain the statevector representing the quantum state from the result.
```

```python
counts=result.get_counts() # Extract the count of measurements obtained from the quantum circuit execution result.
plot_histogram(counts)     # Plot a histogram to visualize the measurement counts.
```

```python
plot_state_qsphere(result.get_statevector()) # Visualize the quantum state using the Q-sphere representation.
```

```python
plot_bloch_multivector(result.get_statevector()) # Plot the quantum state on the Bloch sphere using a multivector representation.
```

<div class="alert alert-block alert-success">
SAMPLE - 2 MULTI QUBIT CIRCUIT
</div>

```python
# importing qiskit package can be done once. If you haven't imported earlier, run this code. 
from qiskit import QuantumRegister, QuantumCircuit, Aer, execute
from qiskit.visualization import *
```

```python
qc1 = QuantumCircuit(3)    # Create a quantum circuit with three qubit.
qc1.x(0)                   # Apply X gate to qubit 0. 
qc1.h(1)                   # Apply a Hadamard gate to qubit 1
qc1.cx(0,1)                # Apply a controlled-X (CNOT) gate with control qubit 0 and target qubit 1.
qc1.cz(1,2)                # Apply a controlled-Z (CZ) gate with control qubit 0 and target qubit 1.
qc1.swap(0,1)              # Swap the states of qubit 0 and qubit 1.
qc1.cswap(0,1,2)           # Apply a controlled-swap (Fredkin) gate with control qubit 0 and qubit 1, and target qubit 2.
qc1.ccx(0,1,2)             # Apply a controlled-controlled-X (Toffoli) gate with control qubits 0 and 1, and target qubit 2.
```

```python
qc1.draw('mpl')
```

```python
job=execute(qc1,backend)   # Execute the quantum circuit 'qc' on the chosen backend
result=job.result()       # Retrieve the result of the executed quantum circuit.
result.get_statevector()  # Obtain the statevector representing the quantum state from the result.
```

```python
counts=result.get_counts() # Extract the count of measurements obtained from the quantum circuit execution result.
plot_histogram(counts)     # Plot a histogram to visualize the measurement counts.
```

```python
plot_state_qsphere(result.get_statevector()) # Visualize the quantum state using the Q-sphere representation.
```

    <b>QUESTION 1.5:<br><br>
        Show that applying the sequence of gates: HZH, to qubit state |0> is equivalent to applying an X-gate<br>
        i.e HZH |0> = X |0> = |1>. Verify using histogram
</b>
</div>

```python
from qiskit import QuantumRegister, QuantumCircuit, Aer, execute
from qiskit.visualization import *

quantum_circuit_5_hzh = QuantumCircuit(1)
#
#
# write your code here
quantum_circuit_5_hzh.h(0)
quantum_circuit_5_hzh.z(0)
quantum_circuit_5_hzh.h(0)

#
#
quantum_circuit_5_hzh.draw('mpl')
```

```python

quantum_circuit_5_x = QuantumCircuit(1)

quantum_circuit_5_x.x(0)

quantum_circuit_5_x.draw('mpl')
```

```python
backend = Aer.get_backend('statevector_simulator')    
job_hzh=execute(quantum_circuit_5_hzh,backend)                   
result_hzh=job_hzh.result()       
result_hzh.get_statevector()  
#
# write code for histogram - verify that probability for occurence of state |1> should be 100 %
plot_histogram(result_hzh.get_counts())
#
#
```

```python
job_x=execute(quantum_circuit_5_x,backend)                   
result_x=job_x.result()       
result_x.get_statevector()  
#
# write code for histogram - verify that probability for occurence of state |1> should be 100 %
plot_histogram(result_x.get_counts())
#
#
```

```python
if result_x.get_counts() == result_hzh.get_counts():
    print(result_x.get_counts())
    print(result_hzh.get_counts())
    print("From the above histograms HZH |0> = X |0> = |1>")
```
