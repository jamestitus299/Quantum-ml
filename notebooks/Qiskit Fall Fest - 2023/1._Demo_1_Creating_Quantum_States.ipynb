# Creating our first Quantum State

# Let's start with the simplest quantum system:

|Œ®‚ü©=|0‚ü©

```python
from qiskit import QuantumRegister, QuantumCircuit, Aer, execute
```

```python
qc = QuantumCircuit(1)  # Create a quantum circuit with one qubit
initial_state = [1,0]   # Define initial_state as |0>
qc.initialize(initial_state, 0) # Apply initialisation operation to the 0th qubit
qc.draw()  # Let's view our circuit
```

```python
backend = Aer.get_backend('statevector_simulator') # Tell Qiskit how to simulate our circuit
```

```python
result = execute(qc,backend).result()
```

```python
out_state = result.get_statevector()
print(out_state) # Display the output state vector
```

```python
from qiskit.visualization import plot_histogram
result = execute(qc,backend).result()
counts = result.get_counts()
plot_histogram(counts) #Display the output in the form of Histogram
```

Applying X Gate to our qubit

```python
from qiskit import QuantumRegister, QuantumCircuit, Aer, execute

#q = QuantumRegister(1)
#hello_qubit = QuantumCircuit(q)

hello_qubit=QuantumCircuit(1)

#hello_qubit.x(q[0])
hello_qubit.x(0)  # moved to state 1
hello_qubit.x(0) # moved to state 0
job = execute(hello_qubit, backend)
result = job.result()
result.get_statevector()
```

```python
hello_qubit.draw('mpl')
```
